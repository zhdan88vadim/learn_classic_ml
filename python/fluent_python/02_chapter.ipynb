Итоговая таблица-шпаргалка
Запись	Название	Что делает
(x for x in items)	Генераторное выражение	Создает объект-генератор. Элементы вычисляются по требованию (лениво).
[x for x in items]	Списковое включение	Создает список (list). Все элементы вычисляются сразу и хранятся в памяти.
{x for x in items}	Множественное включение	Создает множество (set).
{x: y for x,y in items}	Словарное включение	Создает словарь (dict).

The Sequence Comparison Algorithm (Under the Hood)
Python's tuple comparison follows this pseudo-code:

python
def tuple_equal(t1, t2):
    # 1. Same type?
    if type(t1) != type(t2):
        return False
    
    # 2. Same length?
    if len(t1) != len(t2):
        return False
    
    # 3. Compare each element
    for item1, item2 in zip(t1, t2):
        if item1 != item2:
            return False
    
    # 4. All elements matched
    return True

Pitfall 3: == vs is
python
a = (10, 'alpha', [1, 2])
b = a  # b references the SAME object

a == b  # True (same content)
a is b  # True (same object!)

c = (10, 'alpha', [1, 2])  # New tuple
a == c  # True (same content)
a is c  # False (different objects)
