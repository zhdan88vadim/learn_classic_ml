```python
def longest_common_substring_simple(a, b):
    """
    Find the longest common substring using brute force with Python's 'in' operator
    Time: O(m*n^2) in worst case
    """
    # Make sure a is the shorter string
    if len(a) > len(b):
        a, b = b, a
    
    max_length = 0
    longest_substring = ""
    
    for i in range(len(a)):
        for j in range(i + 1, len(a) + 1):
            substring = a[i:j]
            if substring in b:
                print(i, j)
                
                if len(substring) > max_length:
                    max_length = len(substring)
                    longest_substring = substring
    
    return longest_substring, max_length


# Test the simple version
a = 'i have this book i like to read'
b = 'she has this book i like to watch'

result, length = longest_common_substring_simple(a, b)
print(f"Longest common substring: '{result}'")
print(f"Length: {length}")
```

```python
def longest_common_substring_dp(a, b):
    """
    Find the longest common substring using Dynamic Programming
    Time: O(m*n), Space: O(m*n)
    """
    m, n = len(a), len(b)
    
    # Create DP table
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    print(dp)

    max_length = 0
    end_position = 0  # end position in string a
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if a[i-1] == b[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
                if dp[i][j] > max_length:
                    max_length = dp[i][j]
                    end_position = i
            else:
                dp[i][j] = 0
    
    import pandas as pd

    # Convert to DataFrame for beautiful display
    df = pd.DataFrame(dp)
    print(df.to_string())

    # Extract the longest common substring
    longest_substring = a[end_position - max_length:end_position]
    return longest_substring, max_length


# Test the function
a = 'i have this book i like to read'
b = 'she has this book i like to watch'

a = '8861234'
b = '0912345'

result, length = longest_common_substring_dp(a, b)
print(f"Longest common substring: '{result}'")
print(f"Length: {length}")
```

```python
def longest_common_substring_optimized(a, b):
    m, n = len(a), len(b)
    prev = [0] * (n + 1)
    curr = [0] * (n + 1)
    max_length = 0
    end_position = 0
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if a[i-1] == b[j-1]:
                curr[j] = prev[j-1] + 1
                if curr[j] > max_length:
                    max_length = curr[j]
                    end_position = i
            else:
                curr[j] = 0
        prev, curr = curr, prev  # Swap rows
    
    return a[end_position - max_length:end_position], max_length



# Test the function
a = 'i have this book i like to read'
b = 'she has this book i like to watch'

a = '8861234'
b = '0912345'

result, length = longest_common_substring_optimized(a, b)
print(f"Longest common substring: '{result}'")
print(f"Length: {length}")
```

```python
def longest_common_substring_suffix(a, b):
    """
    Find longest common substring using suffix array approach
    """
    # Add unique separators
    combined = a + '#' + b + '$'
    n = len(combined)
    
    # Build suffix array using built-in sort (simple version)
    suffixes = sorted(range(n), key=lambda i: combined[i:])

    print(suffixes)
    
    max_length = 0
    longest_substring = ""
    
    for i in range(n - 1):
        idx1 = suffixes[i]
        idx2 = suffixes[i + 1]
        
        # Check if suffixes come from different strings
        if (idx1 < len(a) and idx2 > len(a)) or (idx2 < len(a) and idx1 > len(a)):
            # Find common prefix length
            common_len = 0
            while (idx1 + common_len < len(a) and 
                   idx2 + common_len < len(combined) and
                   combined[idx1 + common_len] == combined[idx2 + common_len] and
                   combined[idx1 + common_len] != '#' and 
                   combined[idx1 + common_len] != '$'):
                common_len += 1
            
            if common_len > max_length:
                max_length = common_len
                longest_substring = combined[idx1:idx1 + common_len]
    
    return longest_substring, max_length


# Test the suffix array version
a = 'i have this book i like to read'
b = 'she has this book i like to watch'

result, length = longest_common_substring_suffix(a, b)
print(f"Longest common substring: '{result}'")
print(f"Length: {length}")
```
