```python
import numpy as np

# ============================================
# ПРИМЕР 1: Простой случай
# ============================================

# Истинные метки
true_labels = np.array([0, 1, 2, 3, 4, 5, 6])
# Предсказанные метки
pred_labels = np.array([0, 1, 2, 3, 4, 5, 6])

print('-'*30)
print(np.where((true_labels != pred_labels)))
print('-'*30)


misclassified = np.where((true_labels != pred_labels))[0]
print(f"Ошибок нет: {misclassified}")  # []

# ============================================
# ПРИМЕР 2: Есть ошибки
# ============================================

true_labels = np.array([0, 1, 2, 3, 4, 5, 6])
pred_labels = np.array([0, 1, 2, 3, 9, 5, 9])

misclassified = np.where((true_labels != pred_labels))[0]
print(f"Индексы ошибок: {misclassified}")  # [4, 6]

# ============================================
# ПРИМЕР 3: С буквами
# ============================================

class_names = ['А', 'Б', 'В', 'Г', 'Д', 'Е', 'Ё']

true_labels = np.array([0, 1, 2, 3, 4, 5, 6])  # А, Б, В, Г, Д, Е, Ё
pred_labels = np.array([0, 1, 2, 3, 4, 5, 6])  # Все правильно

print("Правильно распознано!")

# А теперь с ошибками
true_labels = np.array([0, 1, 2, 3, 4, 5, 6])
pred_labels = np.array([0, 1, 2, 3, 9, 5, 9])  # Д и Ё перепутаны

misclassified_idx = np.where((true_labels != pred_labels))[0]

print(f"\nНайдено {len(misclassified_idx)} ошибок:")
for idx in misclassified_idx:
    true_letter = class_names[true_labels[idx]]
    pred_letter = class_names[pred_labels[idx]]
    print(f"  Индекс {idx}: {true_letter} → {pred_letter}")
```
