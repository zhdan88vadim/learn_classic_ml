```python
import math

def calculate_psi(expected, actual):
    # expected и actual — это списки долей (например, [0.5, 0.3, 0.2])
    # Добавляем 0.0001, чтобы избежать деления на ноль (сглаживание)
    
    epsilon = 0.0001
    psi_value = 0

    for exp, act in zip(expected, actual):
        exp = exp + epsilon
        act = act + epsilon

        psi_component = (act - exp) * math.log(act/exp)
        psi_value += psi_component

    return psi_value


expected = [0.50, 0.30, 0.20]
actual = [0.70, 0.20, 0.10]

print(round(calculate_psi(expected, actual), 3))  # Выведет: 0.177
```
